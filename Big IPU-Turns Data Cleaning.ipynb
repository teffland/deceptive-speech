### Convert names/fnames to pids and turns/ipus.  Reindex, sort, and drop fname

```python
def name_to_person_ipu(row, name='name'):
    """ Convert p220p221-part1_ch1_2820 to (pid, ipu_id) eg(221, 2820)
    
    Or in the case of turns/ipu files, just return pid
    """
    val = row[name]
    p, l = val.split('-') # p220p221, part1_ch1_2820
    people = p.split('p')  # (,220, 221)
#     print people
    lsplit = l.split('_')  # (part1, ch1, 2820)
    if len(lsplit) == 2:
        part, ch = lsplit
        ipu_or_turn_id = 0
    else:
        part, ch, ipu_or_turn_id = lsplit # part1, _, 2820
        ipu_or_turn_id = int(ipu_or_turn_id) - 1 # convert to int and -1 to index from 0
#     if part == 'part1': # if part one, speaker is the second one
#         person = int(people[2])
#     elif part == 'part2':
#         print'part2'
#         person = int(people[1])
    if ch == 'ch1': # if part one, speaker is the second one
        person = int(people[2])
    elif ch == 'ch2':
#         print'part2'
        person = int(people[1])
    else:
        print "fname formatting wrong... NOT GOOD: ", name

    return person, ipu_or_turn_id # make turns index from 0
```

```python
ipu_dfs = [dal_ipu_train, dal_ipu_dev, dal_ipu_test, 
            liwc_ipu_train, liwc_ipu_dev, liwc_ipu_test, 
            is13_ipu_train, is13_ipu_dev, is13_ipu_test]
turn_dfs = [dal_turns_train, dal_turns_dev, dal_turns_test, 
            liwc_turns_train, liwc_turns_dev, liwc_turns_test, 
            is13_turns_train, is13_turns_dev, is13_turns_test]
turns = [turns_train, turns_dev, turns_test]
ipus = [ipus_train, ipus_dev, ipus_test]
```

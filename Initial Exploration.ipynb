```python
from sklearn.grid_search import GridSearchCV # use as last wrapper for pipelines to do nested CV
from sklearn.cross_validation import cross_val_score, StratifiedKFold
from sklearn.metrics import get_scorer, make_scorer, accuracy_score, auc, f1_score
```

```python
def plot_scores(scores, labels=None, baseline=baseline_accuracy):
    """ Helper function to plot cross validated accuracy scores
    
    Can take a list of scores or a list of a lists of scores
    
    Plots this against naive baseline
    """
    # determine endpoint of baseline
    if type(scores[0]) in [list, np.array, np.ndarray]:
        end = len(scores) + 1
    else:
        end = 2
    fig, ax = plt.subplots(figsize=(12,9))
    ax.boxplot(scores)
    ax.hlines([baseline], [0], [end], linestyles='dashed', label='Baseline Accuracy')
    if labels:
        ax.set_xticklabels(labels)
    plt.legend()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier as RF

rf_param_grid = dict(n_estimators=[10, 25], 
                  criterion=['entropy'], 
                  max_depth=[None, np.sqrt(len(xcols))], 
                  min_samples_split=[2, 5], 
#                   min_samples_leaf=[1], 
#                   min_weight_fraction_leaf=[0.0], 
                  max_features=['sqrt', None], 
#                   max_leaf_nodes=[None], 
#                   bootstrap=[True], 
#                   oob_score=[False], 
#                   n_jobs=[-1], 
#                   random_state=[None], 
#                   verbose=[0], 
#                   warm_start=[False],
#                   class_weight=[None]
                    )
rf = RF()
random_forest = GridSearchCV(estimator=rf, param_grid=rf_param_grid, cv=5, n_jobs=-1, verbose=1)
```

```python
rf_scores, rf_params = nested_cross_val_score(random_forest, X, y, cv=10, verbose=2)
```

```python
print "Crossval accuracy mean:%0.3f, std:%0.3f" % (np.mean(rf_scores), np.std(rf_scores))
print "Crossval selected params: %r" % rf_params
```

## First step is to give all of the tables compatible indices

Note we will treat NEO separately since it has a totally different structure than the other 4.

Each table will get 2 additional columns: 
    - `person` that is the person#
    - `turn` that is the turn#, and will be the index
    

    

```python
def name_to_person_turn(row, name='name'):
    """ Convert p220p221-part1_ch1_2820 to 220, 2820
    
    Only works for DAL, IS13, and LIWC
    Now also handles turns
    """
    name = row[name]
    p, l = name.split('-') # p220p221, part1_ch1_2820
    people = p.split('p')
    lsplit = l.split('_')
    if len(lsplit) == 2:
        ch = lsplit[1]
        turn = '0' # throw away in Turns case
    else:
        _, ch, turn = lsplit # ch1, 2820
    
    if ch == 'ch1':
        person = people[2]
    else:
        person = people[1]
    person_turn = person + '_' + turn
    return person_turn, int(person), int(turn) -1 # make turns index from 0
#     return person_turn
# dal.apply(lambda row: name_to_person_turn(row), axis=1)
```

```python
# dal['person_turn'] = ""; dal['person'] = ""; dal['turn'] = "";
# dal['person_turn'] = dal.apply(lambda row: name_to_person_turn(row)[0], axis=1)
dal['person'] = dal.apply(lambda row: name_to_person_turn(row)[1], axis=1)
dal['turn'] = dal.apply(lambda row: name_to_person_turn(row)[2], axis=1)
dal = dal.set_index('turn')
dal.drop('name', axis=1, inplace=True)
dal.head()
```

```python
is13['person'] = is13.apply(lambda row: name_to_person_turn(row)[1], axis=1)
is13['turn'] = is13.apply(lambda row: name_to_person_turn(row)[2], axis=1)
is13 = is13.set_index('turn')
is13.drop('name', axis=1, inplace=True)
is13.head()
```

```python
liwc['person'] = liwc.apply(lambda row: name_to_person_turn(row)[1], axis=1)
liwc['turn'] = liwc.apply(lambda row: name_to_person_turn(row)[2], axis=1)
liwc = liwc.set_index('turn')
liwc.drop('name', axis=1, inplace=True)
liwc.head()
```

```python
turns['person'] = turns.apply(lambda row: name_to_person_turn(row, name='fname')[1], axis=1)
turns.drop('fname', axis=1, inplace=True)
turns.head()
```

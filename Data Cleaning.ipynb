## First step is to give all of the tables compatible indices

Note we will treat NEO separately since it has a totally different structure than the other 4.

Each table will get 2 additional columns: 
    - `person` that is the person#
    - `turn` that is the turn#, and will be the index
    

    

```python
def name_to_person_turn(row, name='name'):
    """ Convert p220p221-part1_ch1_2820 to 220, 2820
    
    Only works for DAL, IS13, and LIWC
    Now also handles turns
    """
    name = row[name]
    p, l = name.split('-') # p220p221, part1_ch1_2820
    people = p.split('p')
    lsplit = l.split('_')
    if len(lsplit) == 2:
        ch = lsplit[1]
        turn = '0' # throw away in Turns case
    else:
        _, ch, turn = lsplit # ch1, 2820
    
    if ch == 'ch1':
        person = people[2]
    else:
        person = people[1]
    person_turn = person + '_' + turn
    return person_turn, int(person), int(turn) -1 # make turns index from 0
#     return person_turn
# dal.apply(lambda row: name_to_person_turn(row), axis=1)
```

```python
# dal['person_turn'] = ""; dal['person'] = ""; dal['turn'] = "";
# dal['person_turn'] = dal.apply(lambda row: name_to_person_turn(row)[0], axis=1)
dal['person'] = dal.apply(lambda row: name_to_person_turn(row)[1], axis=1)
dal['turn'] = dal.apply(lambda row: name_to_person_turn(row)[2], axis=1)
dal = dal.set_index('turn')
dal.drop('name', axis=1, inplace=True)
dal.head()
```

```python
is13['person'] = is13.apply(lambda row: name_to_person_turn(row)[1], axis=1)
is13['turn'] = is13.apply(lambda row: name_to_person_turn(row)[2], axis=1)
is13 = is13.set_index('turn')
is13.drop('name', axis=1, inplace=True)
is13.head()
```

## standarize the audio features, eg IS13 and Labeled_Turns by person, then remerge and write out to 'normed' files

```python
zscore = lambda x: (x - x.mean()) / x.std()
```

```python
audio_cols = [col for col in turns_test.columns.values if col not in ['label', 'text', 'start', 'end', 'dur', 'person']]
turns_test['maxf0'] = pd.to_numeric(turns_test['maxf0'], errors='coerce')
turns_test_group = turns_test.groupby('person')
turns_test[audio_cols] = turns_test_group[audio_cols].transform(zscore)
# turns_test.head()

# turns_train['maxf0'][~turns_train.applymap(lambda x: isinstance(x, (int, float))).all(1)]
turns_train['maxf0'] = pd.to_numeric(turns_train['maxf0'], errors='coerce') # one random val is a string...
turns_train.fillna(0.0, inplace=True)
turns_train_group = turns_train.groupby('person')
turns_train[audio_cols] = turns_train_group[audio_cols].transform(zscore)
# turns_train.head()
```

```python
audio_cols = [col for col in is13_test.columns.values if col not in ['label', 'text', 'start', 'end', 'dur', 'person']]
is13_test_group = is13_test.groupby('person')
is13_test[audio_cols] = is13_test_group[audio_cols].transform(zscore)
# is13_test.head()

is13_train_group = is13_train.groupby('person')
is13_train[audio_cols] = is13_train_group[audio_cols].transform(zscore)
# is13_train.head()
```

```python
# normalize DAL and LIWC also
audio_cols = [col for col in dal_test.columns.values if col not in ['label', 'text', 'start', 'end', 'dur', 'person']]
dal_test_group = dal_test.groupby('person')
dal_test[audio_cols] = dal_test_group[audio_cols].transform(zscore)
# is13_test.head()

dal_train_group = dal_train.groupby('person')
dal_train[audio_cols] = dal_train_group[audio_cols].transform(zscore)
```

```python
# normalize DAL and LIWC also
audio_cols = [col for col in liwc_test.columns.values if col not in ['label', 'text', 'start', 'end', 'dur', 'person']]
liwc_test_group = liwc_test.groupby('person')
liwc_test[audio_cols] = liwc_test_group[audio_cols].transform(zscore)
# is13_test.head()

liwc_train_group = liwc_train.groupby('person')
liwc_train[audio_cols] = liwc_train_group[audio_cols].transform(zscore)
```
